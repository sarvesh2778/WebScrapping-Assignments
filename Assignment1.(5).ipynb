python program for new updates and make data frame

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.cnbc.com/world/?region=world')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
soup
```

```python
titles=[]
for i in soup.find_all('a',"LatestNews-headline"):
    titles.append(i.text)
    print(titles)
```

```python
time=[]
for i in soup.find_all('time',"LatestNews-timestamp"):
    time.append(i.text)
    print(time)
```

```python
import pandas as pd
df = pd.DataFrame({'titles':titles,'time':time})
```
